# Line Plotting with Matplotlib. 

```python
# Importing the Necessary Packages.
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# This will create numbers from 0 to 4 (five numbers.)
x = np.arange(5)
y = x
print(x)
```

```python
plt.plot(x, y, 'o--')     # This would plot a broken line
#plt.plot(x, y, 'o')      # This would plot just the dot o only.
#plt.plot(x, y, 'o-')   # This would join all the lines together. 
plt.show() 
```

```python
# creating a range of values from lower-boundary 1 to upperboundary 10.
# and the numbers created should be 10 in number. 
N = 10 
x = np.linspace(1, 10, N)
print(x)
y = x 
```

```python
plt.plot(x, y, 'o--')
plt.grid(True)
plt.show()
```

```python
# Finding the Log using numpy and plotting a graph  
y = np.logspace(0.1, 1, N)
print(y)
plt.plot(x, y, 'o--')
plt.grid(True)
plt.show() 
```

## Multiline Plotting with matplotlib 

```python
x = np.arange(10)
plt.title('Multiline Plotting.')
plt.plot(x, x**2)
plt.plot(x, x**4)
plt.plot(x, 4**x)
plt.plot(x, x**3)
plt.grid(True)
plt.show() 
```

## Grids And Axis 

```python
# Setting the grid, and showing the boundary of the x and y axis. 
x = np.arange(3)
plt.plot(x, x**2, x, x**3)
plt.grid(True)
plt.axis([0, 2, -1, 8.5])
print(plt.axis())    # This will show us the limit of the x and y axis. 
plt.show() 
```

```python
# Setting the axis of the Graph to be new values. 
x = np.arange(5)
plt.plot(x, x**2)
y = np.sin(x)
plt.grid(True)
plt.plot(x, y, 'o--')
plt.axis([0, 2, 0, 5])   # Setting the x-axis from 0 to 2, and the y-axis form 0 to 5.
plt.show() 
```

```python
N = 10
y = np.logspace(0.1, 1, N)
x = np.arange(10)
print(y, x)
plt.plot(x, y, 'o--')
plt.grid(True)
plt.xlim([0, 10])     # Setting the limit of x-axis to be from 0 to 10. 
plt.ylim([0, 11])     # Setting the limit of y-axis to be from 0 to 11. 
plt.show() 
```

## Adding Labels to the x and y Axis. 

```python
N = 10
y = np.logspace(0.1, 1, N)
x = np.arange(10)
x = np.cos(x)
print(y, x)
plt.plot(x, y, 'o--', 
        label='x vs y')
plt.grid(True)
plt.xlabel('x = cos(x)')   # Setting the label for x-axis 
plt.ylabel('y = np.logspace')  # Setting the label for the y-axis 
plt.xlim([-5, 2])     # Setting the limit of x-axis to be from 0 to 10. 
plt.ylim([0, 11])     # Setting the limit of y-axis to be from 0 to 11. 
plt.title('A Graph of x against y')
plt.legend(loc='upper left')
plt.show() 
```

## Adding label to the Graph 

```python
# Setting the axis of the Graph to be new values. 
x = np.arange(5)
plt.plot(x, x**2, 
         label='x**2')
y = np.sin(x)
plt.grid(True)
plt.plot(x, y, 'o--',
        label='x vs y')
plt.axis([0, 2, 0, 5])   # Setting the x-axis from 0 to 2, and the y-axis form 0 to 5.
plt.legend(loc='upper left') 
plt.show() 
```

```python
# Plotting A pi-chart. 
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 120]
colors = ['gold', 'green', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
```

```python
# Plotting the Graph 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.grid(True)
plt.savefig('piechart.jpg')
plt.show() 
```

```python
# Plotting a sine wave Graph. 
value = np.arange(0, 3*np.pi, 0.1)
y = np.sin(value)
z = np.cos(value)
```

```python
# Plotting the Graph 
plt.title('Sine wave form Graph')
plt.plot(value, y)
plt.grid(True)
plt.show() 
```

```python
plt.plot(value, z, marker="x")
plt.grid(True)
plt.show() 
```

```python
# Generate a sequence of numbers from -10 to 10 with 100 steps in between
x = np.linspace(-10, 10, 100)
# Create a second array using sine
y = np.sin(x)

# The plot function makes a line chart of one array against another
plt.plot(x, y, 'o--', marker="x", label='Sine-wave-Function')
plt.legend(loc='upper left')
plt.grid(True)
plt.xlabel('x= np.linspace')
plt.ylabel('y = np.sin(x)')
plt.xlim([-10.0, 10.0])
plt.ylim([-1.0, 2.0])
plt.title('Sine Wave Graph.')
plt.show() 
```

```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = plt.imread('nedu.jpg')
img = np.rot90(img)
```

```python
plt.imshow(img) 
print(img.shape)
print(img.size, img.dtype)
plt.title('Nedu Picture')
plt.show()
```

```python
new_img = np.roll(img, 2500)
plt.imshow(np.roll(img, 256)) 
print(img.shape)
print(img.size, img.dtype)
plt.title('Nedu Picture')
plt.imsave('new.png', new_img)
plt.show()
```

```python
r = img[:, :, 0]
g = img[:, :, 1]
b = img[:, :, 2]

output = [img, r, g, b]

titles = ['Image', 'Red', 'Green', 'Blue']
for i in range(4):
    plt.subplot(2, 2, i+1)
    
    plt.axis('off')
    plt.title(titles[i])
    if i == 0:
        plt.imshow(output[i])
        
    else:
        plt.imshow(output[i], cmap='gray')
        
plt.show()         
```
